```python
import matplotlib.pyplot as plt
import numpy as np
```

# Compute the Harmonic Sum using Single Precision

```python
# Calculate and plot on the graph 
maxN = 9
minN = 1
n = range(minN, maxN)
```

```python
# Define needed functions
def Sum_Single_Precision(n):
    sum = 0
    for k in range(1, n):
        sum += 1/k
    return format(sum, '.1f')

for i in n:
    N = 10**i
    SumSingle = Sum_Single_Precision(N)
    plt.plot(N, SumSingle, 'bo')

plt.xlabel('N')
plt.ylabel('Partial Sums')
plt.show()
```

# Compute the Harmonic Sum using Double Precision

```python
def Sum_Double_Precision(n):
    sum = 0
    for k in range(1, n):
        sum += 1/k
    return format(sum, '.2f')

for i in range(minN, maxN):
    n = 10**i
    SumDouble = Sum_Double_Precision(n)
    plt.plot(n, SumDouble, 'ro')

plt.xlabel('N')
plt.ylabel('Partial Sums')
plt.show()
```
