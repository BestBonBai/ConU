```python
import numpy as np



def main_function(x):
    return np.sin(np.pi*x)

exact_value = 2/(np.pi)
```

## Approximating the integration using *Midpoint Rule*

```python
def x(k, h):
    return k*h

def Midpoint(M, k, h):
    # Terminate point
    if k == 0:
        return 0

    X_k = x(k, h) 

    X_k = X_k + h/2
    return main_function(X_k) + Midpoint(M, k-1, h)

def MP_approximate(M):
    h = 1/M
    return h * Midpoint(M, M, h)
 
        
```

```python
noOfM = range(1, 8)
result = []
M_range = []
for m in noOfM:
    m = 2**m
    M_range.append(m)
    result.append(MP_approximate(m))

for Sum, m in zip(result, M_range):
    error = abs(exact_value - Sum)
    print('M = ', m, ' || APPROXIMATION: ', Sum, ' || ERROR: ', error)
```
