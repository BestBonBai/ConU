```python
# import modules
import matplotlib.pyplot as plt
import numpy as np
```

### Calculate Sequence: a recursive function used to calculate a member of the sequence with a given value of N

```python
def CalculateSequence(n):
    if n == 0:
        return 1
    return (2*CalculateSequence(n-1)**3 + 5) / (3*CalculateSequence(n-1)**2)
```

### Printing the results after running through the function with each exponential defined in range 0 to N

```python
N = 21

array = range(0, N)
result = map(CalculateSequence, array)
result = list(result)

print(result)
```

### Plot the results on the graph with respect to N

```python
plt.plot(array, result, 'r-')
plt.xlabel("N")
plt.ylabel("Value of x^(n)")

plt.show()
```

### Observed behaviour:
The graph reached its peak when N = 2 with value = 2.33, then stablizes at value 1.709 as N increases.
